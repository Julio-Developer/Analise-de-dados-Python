# Analise de dados Exploratoria

```python
# Importando as blibiotecas que serão usadas

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Definindo o estilo dos gráficos

plt.style.use('seaborn')
```

```python
# Pegando a base de dados

df = pd.read_excel('AdventureWorks.xlsx')
```

```python
df.head(25)
```

```python
# Mostranod os 15 ultimos registros 

df.tail(15)
```

```python
# Verificando o tipo de cada coluna

df.dtypes
```

```python
# Qual foi a receita total?

df['Valor Venda'].sum()
```

```python
# Qual o valor de custo por produto?

# Criando uma nova coluna com os dados do custo toral
df['Custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
df
```

```python
# Custo total

round(df['Custo'].sum(), 2)
```

```python
# Qual o custo total por produto?

# Criando coluna com os dados de custo total

df['Lucro'] = df['Valor Venda'] - df['Custo']
```

```python
df
```

```python
# Lucro total

round(df['Lucro'].sum(),2)
```

```python
# Qual foi a  quantidade de dias par enviar o produtos? 

df['Tempo Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df
```

```python
# Alterando a coluna Tempo de Envio para data

df['Tempo Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(15)
```

```python
df.groupby('Marca')['Tempo Envio'].mean()
```

```python
# Verificando se há valores ausentes

df.isnull().sum()
```

```python
# Ajustando os valores de notação para padrão dolar

pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Verificando o lucro por ano e por marca

df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
# Transformando em data freme o agrupamento acima

lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
```

```python
lucro_ano
```

```python
# Qual o total de produtos vendido?

df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Transformando em grafico os dados acima

df.groupby('Produto')['Quantidade'].sum().plot.barh(title='Total de Produtos Vendidos')
plt.xlabel('Produtos')
plt.ylabel('Quatidade Vendida');
```

```python
# Lucro por ano

df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('REceita');
```

```python
# Analisando as vendas de um ano especifico

df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head(10)
```

```python
# Lucro por mês do ano de 2009

df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title="Lucro x Mês")
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
# Em 2009, qual a marca que trouxe mais lucro?

df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title='Lucro por marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
# Lucro por classe

df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title='Classe x Lucro')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
